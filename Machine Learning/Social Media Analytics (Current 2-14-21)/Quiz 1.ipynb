```python
import networkx as nx
G = nx.read_edgelist(r"C:\Users\cluer\OneDrive - Virginia Tech\Classes\Social Media Analytics\CA-GrQc.txt")
```

```python
# Question 1
# What is the average node degree of the ca-GrQc network?
    # 2.76
    # 5242
    # 14496
    # 5.53

# Average degree is simply the average number of edges per node in the graph. It is relatively straightforward to calculate.
# Total Edges/Total Nodes=Average Degree

nx.Graph.number_of_edges(G)/nx.Graph.number_of_nodes(G)

# Answer: 2.76
```

```python
# Question 2
# Which are the most central nodes in ca-GrQc based on degree and closeness centrality, respectively?
# Hint: the following code sorts the nodes based on their centrality values (descending order)

dc = nx.degree_centrality(G)
import operator
sorted_dc = sorted(dc.items(), key=operator.itemgetter(1),reverse=True)

    # 21012, 13801
    # 13801, 21012
    # 14485, 21281
    # 21281, 14485

print(sorted_dc)

# Answer: 21012, 13801
```

```python
# Question 3
# How many connected components are there in the ca-GrQc network?
    
nx.number_connected_components(G)

# Answer: 335
```

```python
# Question 4
# What is the size (in # of nodes) of the giant connected component in ca-GrQc and its density?
# Hint: You can extract the giant connected component using the following code snippet:

giant_nodes = max(nx.connected_components(G), key=len)
giant_graph = G.subgraph(giant_nodes).copy()

    # 4158, 0.001
    # 13428, 0.02
    # 5242, 0.001
    
nx.number_of_nodes(giant_graph)

# Answer: 4158, 0.001
```
